```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/Sales_Analysis.txt",sep = ",",header = None,
                 names=("OrderID","Date","Region","Category","Product","Sales","Profit","Quantity"))
df
```

```python
# Plot a line chart to show the monthly sales trend.

df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month
monthly_sales = df.groupby("Month")["Sales"].sum()
plt.plot(monthly_sales.index,monthly_sales.values,marker = "o",ms = 15,mec = "m",mfc = "b",color = "g")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Line Chart")
plt.show()
```

```python
# Plot a bar chart to display total sales by product category

total_sales = df.groupby("Category")["Sales"].sum()
plt.bar(total_sales.index,total_sales.values)
plt.xlabel("Category")
plt.ylabel("Sales")
plt.title("Total Category Sales Bar Chart")
plt.show()
```

```python
# Plot a horizontal bar chart to visualize total profit by region.

total_profit = df.groupby("Region")["Profit"].sum().sort_values()
plt.barh(total_profit.index,total_profit.values)
plt.xlabel("Profit")
plt.ylabel("Region")
plt.title("Total Profit by Region")
plt.show()
```

```python
# Plot a histogram to analyze the distribution of sales values.

plt.hist(df["Sales"],bins = 5)
plt.xlabel("Sales Values")
plt.ylabel("Frequency")
plt.title("Histogram-Sales Values")
```

```python
# Plot a scatter plot to show the relationship between sales and profit.

plt.scatter(df["Sales"],df["Profit"])
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Relationship between Sales and Profit")
plt.show()
```

```python
# Plot a pie chart to represent the percentage contribution of sales by category.

total_sales = df.groupby("Category")["Sales"].sum()
plt.figure(figsize=(15,10))
plt.pie(total_sales.values,labels = total_sales.index, autopct="%1.2f%%")
plt.title("Percentage contribution of Sales by Category")
plt.legend(loc = "upper right")
plt.show()
```

```python
# Plot a bar chart to show total quantity sold for each product.

total_quantity = df.groupby("Product")["Quantity"].sum()
plt.bar(total_quantity.index,total_quantity.values)
plt.xticks(rotation = 90)
plt.show()
```

```python
# Plot a bar chart to identify the top 5 products based on total sales.

total_sales = df.groupby("Product")["Sales"].sum().sort_values(ascending=False).head(5)
plt.bar(total_sales.index,total_sales.values)
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Top 5 products")
plt.show()
```
